# 1. Importing Modules

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

# 2. Loading the iris data

```python
dataset = load_iris()
```

# 3. Summary of Data

```python
dataset.data
```

```python
dataset.data.shape
```

```python
print('Classes to predict: ', dataset.target_names)
```

```python
print('Features: ', dataset.feature_names)
```

# 4. Extracting data attributes / features & Extracting target/ class labels

```python
X = dataset.data
y = dataset.target
```

```python
display(X.shape, y.shape)
```

# 5. Importing Scikit-Learn library

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

# 6. Splitting the dataset into train and test.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 50, test_size = 0.25)
```

# 7. Fit the classifier

```python
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

# 8. Prediction

```python
y_pred = clf.predict(X_test)
```

# 9. Check Accuracy

```python
print('Accuracy Score on train data (using Default criterionas gini): ', accuracy_score(y_true=y_train, y_pred=clf.predict(X_train)))
print('Accuracy Score on test data (using Default criterionas gini): ', accuracy_score(y_true=y_test, y_pred=y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred) 
```

# 10. Visualize the Tree

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(clf, feature_names = dataset.feature_names, class_names = dataset.target_names, rounded = True, filled = True);
```
