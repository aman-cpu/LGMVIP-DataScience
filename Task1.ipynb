# 1. Importing Modules

```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

# 2. Loading and Plotting Original Image

```python
img = cv2.imread("image.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(7,7))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

# 3. Converting Image to GrayScale

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(7,7))
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

# 4. Inverting the Image

```python
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(7,7))
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(7,7))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

# 5. Converting your images to pencil sketches

```python
final = cv2.divide(img_gray, 255 - img_smoothing, scale=255)
plt.figure(figsize=(7,7))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Pencil Sketch Image")
plt.show()
```

# 6. The Final Output

```python
plt.figure(figsize=(20,20))
plt.subplot(1,5,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,5,4)
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.subplot(1,5,5)
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Pencil Sketch Image")
plt.show()
```
